```python
#1.Sigmoid
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#sigmoid fn
def sigmoid(x):return 1/(1+np.exp(-x))

#input value
x=np.linspace(-10,10,100) #generate 100 values bw -10 to 10
y=sigmoid(x)

#plot
plt.plot(x,y)   #line graph -> x axis we have input values and y axis is the sigmoid fn
plt.title("S(sigmoide) curve")
plt.xlabel('x')
plt.ylabel('Ïƒ (x)')
plt.grid()
plt.show()

#fastest learning -> point five
#we should avoid both initial and final saturation
```

```python
#2. ReLU

#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#ReLU
def ReLU(x): return np.maximum(0,x)

#input value
x=np.linspace(-10,10,100) #generate 100 values bw -10 to 10
y=ReLU(x)

#plot
plt.plot(x,y)   #line graph x axis we have input values and y axis is the relu fn
plt.title("ReLU curve")
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
