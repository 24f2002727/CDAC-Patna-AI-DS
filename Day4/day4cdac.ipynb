```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
```

```python
#creating raw datasets
data={
    'Size':[1200,1500,1000,2000],
    'Bedrooms':[3,4,np.nan,5],
    'Location':['urban','suburban','urban','rural'],
    'Price':[3000000,400000,250000,5000000]
}

#we will be storing the data to tables
df=pd.DataFrame(data)

#updatng the missing data before normalisation
df['Bedrooms'].fillna(df['Bedrooms'].mean(),inplace=True)


```

```python
print(df)

#now we will be encoding the categorical data
encoder=LabelEncoder()
df['Location']=encoder.fit_transform(df['Location']) #here fit-transform transforms the string value to no and also remembers it.
print(df)

#normalisation technique
scaler=MinMaxScaler()
df[['Size','Bedrooms']]=scaler.fit_transform(df[['Size','Bedrooms']])
print(df)
```
