# **Introduction to Data Visualization**

**What is Data Visualization ?**
* Graphical representation of information and data using visual elements like charts, graphs, maps, and dashboards
* Helps transform raw data into a visual context, making it easier to identify patterns, trends, outliers, and insights

**Why It Matters in Data Analysis, Reporting, and Communication ?**
* Simplifies Complex Data
* Reveals Trends & Patterns (e.g. share price)
* Improves Decision-Making
* Enhances Communication

**Real-World Examples**
* Health Dashboards
* Financial Data
* COVID-19 Dashboards


# **Plot Basics & Line Properties**

```python
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

plt.plot(x, y, color='blue', linestyle=':', marker='o', linewidth=2, label='Square')
plt.title("Customized Line Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend()
plt.grid()
plt.show()
# '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
```

# **Visual Design Considerations**

**Objective**: Understand how to make graphs clear, accurate, and useful.
* Simplicity > complexity

* Avoid chartjunk (unnecessary decorations): unnecessary graphics (like 3D effects, heavy shadows, flashy backgrounds)

* Use consistent colors, scales, and labels: e.g., If one bar is blue and another is green without meaning, its confusing

# **Multiple Plots and Annotations**

```python
#  add 3 plots within the same axis.
import numpy as np
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]
y1 = [1, 2, 3, 4]

plt.plot(x, y1, color='orange', marker = 'o' , linewidth=2, label='Linear')

plt.plot(x, np.sin(x), color='green', marker = 'o', linewidth=2, label='Sine')

plt.plot(x, y, color='blue', marker='o', linewidth=2, label='Squared')

plt.title("Customized Line Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend()
plt.show()
```

```python
plt.fill_between(x, y, y1, color='magenta', alpha=0.5)
```

# **Subplots and Layouts**

```python
fig, axs = plt.subplots(2, 2)

x = np.linspace(0, 10, 100)
axs[0, 0].plot(x, np.sin(x), color='orange')
axs[0, 0].set_title("Sine")

axs[0, 1].plot(x, np.cos(x), color='magenta')
axs[0, 1].set_title("Cosine")

axs[1, 0].plot(x, np.exp(x), color='green')
axs[1, 0].set_title("Exponential")

axs[1, 1].plot(x, np.log(x + 1), color='blue')
axs[1, 1].set_title("Logarithm")

fig.suptitle("Subplots Example")
plt.tight_layout()
plt.show()
```

# **Various Chart Types**

```python
# Bar chart
plt.bar(['A', 'B', 'C'], [10, 20, 15])
```

```python
# Histogram
plt.hist(np.random.randn(1000), bins=10)
```

```python
# Scatter plot
plt.scatter(np.random.rand(50), np.random.rand(50))
```

```python
# Box Plot
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
plt.boxplot(data)
```

```python
# Pie Chart

sizes = [20, 30, 50]
labels = ['Group A', 'Group B', 'Group C']
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
```
