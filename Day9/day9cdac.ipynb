```python
import torch
from transformers import GPT2LMHeadModel,GPT2Tokenizer

model_name="distilgpt2"
tokenizer=GPT2Tokenizer.from_pretrained(model_name)
model=GPT2LMHeadModel.from_pretrained(model_name)

device="cuda" if torch.cuda.is_available() else "cpu"  #use gpu if gpu is available use cpu
model.to(device)

#input prompt
prompt="once upon a time in universe"
inputs=tokenizer(prompt,return_tensors="pt").to(device) #here pt is pytorch

#generate text
output=model.generate(
    inputs["input_ids"],
    max_length=50,            #max length 50
    pad_token_id=tokenizer.eos_token_id,      #
    num_return_sequences=1,   #return only one
    no_repeat_ngram_size=2,    #
    do_sample=True,
    top_k=100,       #return only top50
    top_p=.95,        #return only those >.95 accuracy
    temperature=.1  #variation
)

#decode and print
generated_text=tokenizer.decode(output[0],skip_special_tokens=True) #
print(generated_text)
```
